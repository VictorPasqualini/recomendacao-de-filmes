```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# lendo csv movies.csv
movies_columns = ['id_filme', 'titulo', 'genero']
ds_movies = pd.read_csv('movies.csv', header = 0, names = movies_columns)
ds_movies.head()
```

```python
# lendo csv ratings.csv
ratings_columns = ['id_usuario', 'id_filme', "avaliacoes", "timestamp"]
ds_ratings = pd.read_csv('ratings.csv', header = 0, names = ratings_columns)
ds_ratings.head()
```

```python
# merge dataset ds_movies com ds_ratings pelo id_filme
ds_ratings_movies_v1 = pd.merge(ds_ratings, ds_movies, how = 'inner', on = 'id_filme')
ds_ratings_movies_v1.head()
```

```python
# novo dataset com colunas agrupadas por titulo
ds_count_ratings_v1 = ds_ratings_movies_v1.groupby('titulo')['avaliacoes'].count().reset_index()
ds_count_ratings_v1
```

```python
# merge dataset ds_rating_movies_v1 com ds_count_ratings_v1 pelo titulo
ds_count_ratings_final = ds_ratings_movies_v1.merge(ds_count_ratings_v1, on = 'titulo')
ds_count_ratings_final
```

```python
# novo dataset apenas com registros que tenham quantidade de avaliacoes maior ou igual a 20
ds_ratings_movies_final = ds_count_ratings_final[ds_count_ratings_final['avaliacoes_y'] >= 20]
ds_ratings_movies_final.shape
```

```python
# renomeando a coluna avaliacoes_x para avaliacao e a coluna avaliacoes_y para quantidade de avaliacoes
ds_ratings_movies_final.rename(columns = {'avaliacoes_x': 'avaliacao', 'avaliacoes_y': 'quantidade de avaliacoes'}, inplace = True)
ds_ratings_movies_final
```

```python
ds_ratings_movies_final.drop_duplicates(['id_usuario', 'titulo'], inplace = True)
ds_ratings_movies_final
```

```python
pivot_movies = ds_ratings_movies_final.pivot_table(columns = 'titulo', index = 'id_usuario', values = 'avaliacao')
pivot_movies
```

```python
pivot_movies = pivot_movies.fillna(0)
pivot_movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import operator
```

```python
def similar_users(user_id, movies, k = 5):

    # cria um dataframe apenas para o usuario que se recomendará os filmes
    df_user = movies[movies.index == user_id]

    # cria um datafram para os outros usuarios
    df_other_users = movies[movies.index != user_id]

    # calcula a similaridade do cosseno entre o usuario desejado e todos os outros usuarios
    similarities = cosine_similarity(df_user, df_other_users)[0].tolist()

    # cria indice para os outros usuarios
    index = df_other_users.index.tolist()

    # cria um dicionario com os pares dos usuarios e os indices de similaridades
    index_similarity = dict(zip(index, similarities))

    # ordena por similaridade
    index_similarity_sorted = sorted(index_similarity.items(), key = operator.itemgetter(1))
    index_similarity_sorted.reverse()

    # coloca os usuarios mais similares no topo e os retorna
    top_n_users_similarities = index_similarity_sorted[:k]
    users = [u[0] for u in top_n_users_similarities]

    return users

current_user = 1

# pega os usuarios mais similares ao usuario desejado
similar_user_index = similar_users(current_user, pivot_movies)
print(similar_user_index)
```

```python
def recommend_item(user_index, similar_user_index, movies, items = 10):

    # carrega o vetor com usuarios similares
    similar_users = movies[movies.index.isin(similar_user_index)]

    # calcula a media dos itens usando os usuarios similares
    similar_users = similar_users.mean(axis = 0)

    # converte em dataframe para facilitar a filtragem
    df_similar_users = pd.DataFrame(similar_users, columns = ['media'])


    # carrega o vetor para o usuario que se deseja recomendar o filme
    df_user = movies[movies.index == user_index]

    # transpor esse dataframe para facilitar a filtragem
    df_user_transposed = df_user.transpose()

    # renomeia a coluna para avaliacao
    df_user_transposed.columns = ['avaliacao']

    # remove linhas com avaliacoes 0 (filmes nao avaliados pelos usuarios)
    df_user_transposed = df_user_transposed[df_user_transposed['avaliacao'] == 0]

    # gera lista de filmes que o usuario ainda nao avaliou
    unvisited_movies = df_user_transposed.index.tolist()


    # filtra a media das avaliacoes dos usuarios similares para os filmes ainda nao visitados
    df_similar_users_filtered = df_similar_users[df_similar_users.index.isin(unvisited_movies)]

    # ordena os filmes com as melhores notas
    df_similar_users_ordered = df_similar_users_filtered.sort_values(by = ['media'], ascending = False)

    # pega os filmes com as melhores notas
    top_n_hotels = df_similar_users_ordered.head(items)

    return top_n_hotels

df_recommend = recommend_item(current_user, similar_user_index, pivot_movies)
df_recommend
```
